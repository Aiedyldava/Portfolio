# U.S. Medical Insurance Costs

```python
import csv
```

* find average age of the patients
* return the number of males vs. females counted in the dataset
* find geographical location of the patients
* return the average yearly medical charges of the patients
* creating a dictionary that contains all patient information

```python
# create the lists
name = []
age = []
sex = []
bmi = []
children = []
smoker = []
region = []
charges = []
insurance_data = []
```

```python
# import all data from csv to existed list
with open('insurance.csv','r') as insurance_data:
    reader = csv.DictReader(insurance_data)
    for x in reader:
        age.append(x['age'])
        sex.append(x['sex'])
        bmi.append(x['bmi'])
        children.append(x['children'])
        smoker.append(x['smoker'])
        region.append(x['region'])
        charges.append(x['charges'])
        
```

```python
# total patient
total_patient = len(age)
print("There are {} patients.".format(total_patient))
```

```python
# find average age of patient
def age_analytics():
    average_age = 0

    for temp in age:
        average_age += int(temp)

    average_age /= total_patient

    print('The average of patient age is {}'.format(format(average_age, '.2f')))

age_analytics()
```

```python
# male vs female data
def sex_analytics():
    males = 0
    females = 0

    for temp in sex:
        if temp == 'male':
            males += 1
        elif temp == 'female':
            females += 1

    print('There are {} males and {} females patient'.format(males,females))

sex_analytics()
```

```python
# find geographical location of the patients
def region_analytics():
    region_list = []
    region_list_num = []
    for temp in region:
        if temp not in region_list:
            region_list.append(temp)
        else:
            continue

    for x in range(len(region_list)):
        temp = 0
        for y in region:
            if y == region_list[x]:
                temp += 1
        region_list_num.append(temp)

    region_dict = {key:value for key,value in zip(region_list,region_list_num)}  
    print('There are {} region with:'.format(len(region_list)))
    for x,y in region_dict.items():
        print('{y} people\'s lived in {x}'.format(x = x.title(),y = y))

region_analytics()
```

```python
# average yearly medical charges of the patients
def charge_analytics():
    average_charge = 0
    total_charge = len(charges)
    for temp in charges:
        average_charge += float(temp)

    average_charge /= total_charge

    print('The average of medical bills charges is {}'.format(format(average_charge, '.2f')))
    
charge_analytics()
```

```python
# create mock name for patient
for x in range(total_patient):
    temp = "Person " + str(x+1)
    name.append(temp)
```

```python
# dictionary that contains all patient information
all_patient_data = {}

for temp in range(total_patient):
    all_patient_data[name[temp]] = {"Age": age[temp],"Sex": sex[temp],"Bmi": bmi[temp],"Children": children[temp],"Smoker": smoker[temp],"Region": region[temp],"Charges": charges[temp]}
```

```python
# print all data
message = """{}
Age: {}, Sex: {}, Bmi: {}, Children: {}, Smoker: {}, Region: {}, Charges: {}
"""
for name,record in all_patient_data.items():
    print(message.format(name,record['Age'],record['Sex'],record['Bmi'],record['Children'],record['Smoker'],record['Region'],record['Charges']))
```
